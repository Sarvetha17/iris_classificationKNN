```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report,confusion_matrix
```

```python
df= pd.read_csv("iris flower classification.csv")
df
```

```python
df.info()
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
x=df[['sepal_length','sepal_width','petal_length','petal_width']]
y=df[['species']]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=42)
```

```python
doc1 = KNeighborsClassifier(n_neighbors=11,metric = "minkowski",p=5)
doc1.fit(x_train,y_train)
```

```python
y_pred=doc1.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
doc2 = accuracy_score(y_test,y_pred)
print("Accuracy:",doc2)
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

```python
import pickle
pickle_out=open("classification_KNN.pkl","wb")
pickle.dump(doc1,pickle_out)
pickle_out.close()
```
